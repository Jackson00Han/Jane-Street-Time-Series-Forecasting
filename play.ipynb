```python
# 为每个操作加上注释
def pluralize(word, amount):
    """Return the plural form of word if amount is not 1."""
    # Check for specific cases， 单复数不变
    if word == "sheep":
        plural_word = "sheep"
    # Check for specific cases， 单复数不规则变化
    elif word == "goose":
        if amount == 1:
            plural_word = "goose"
        else:
            plural_word = "geese"
    # Regular pluralization rule 一般规则
    else:
        if amount == 1:
            plural_word = word
        else:
            plural_word = word + "s"
    return plural_word
```
